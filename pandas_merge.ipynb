```python
import pandas as pd
df1 = pd.DataFrame({
    "city": ["new york","chicago","baltimore",'orlando'],
    "temp": [21,14,35,32]
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san"],
    "humid": [65,68,75]
})
df2
```

```python
# it will only merge with 'inner join' or union of both data frames
```

```python
df3 = pd.merge(df1,df2, on = 'city')
df3
```

```python
df3 = pd.merge(df1,df2, on = 'city', how='outer')
df3
```

```python
# left data frame depends on the order in which it is passed as argument | how='inner' is by default
```

```python
df3 = pd.merge(df1,df2, on = 'city', how='left')
df3
```

```python
df3 = pd.merge(df1,df2, on = 'city', how='right')
df3
```

```python
df3 = pd.merge(df1,df2, on = 'city', how='outer', indicator=True)
df3
```

```python
# if the columns are repeating in both data frames, we can use suffixes.default(_x,_y)
```

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35,38],
    "humidity": [65,68,71, 75]
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "temperature": [21,14,35],
    "humidity": [65,68,71]
})
df2
```

```python
df3 = pd.merge(df1,df2, on='city', how='outer', suffixes=('_first','_second'))
df3
```
